### Order Processing Time 

```python
ord_df["order_date"] = pd.to_datetime(ord_df["order_date"])

ship_df["dispatch_date"] = pd.to_datetime(ship_df["dispatch_date"])
ship_df["delivery_date"] = pd.to_datetime(ship_df["delivery_date"])
```

```python
ord_ship = pd.merge(ord_df, ship_df, on="order_id", how="left")
ord_ship
```

```python
ord_ship["processing_time"] = (ord_ship["dispatch_date"] - ord_ship["order_date"]).dt.days

ord_ship
```

```python
ord_ship[["order_id", "order_date", "dispatch_date", "processing_time"]]
```

### On-Time Delivery Rate

```python
ship_df
```

```python
ship_car = pd.merge(ship_df, car_df, on="carrier_id", how="left")
ship_car
```

```python
ship_car = ship_car[ship_car["status"] != "canceled"]

ship_car
```

```python
ship_car["actual_delivery_time"] = (ship_car["delivery_date"] - ship_car["dispatch_date"]).dt.days

ship_car
```

```python
ship_car["on_time"] = ship_car["actual_delivery_time"] <= ship_car["avg_delivery_time"]

ship_car
```

```python
on_time_rate = ship_car["on_time"].mean() * 100

on_time_rate
```

### Carrier Performance
